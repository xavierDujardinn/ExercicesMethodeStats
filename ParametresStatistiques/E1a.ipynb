```python
import pandas
import matplotlib.pyplot as plt

input_df = pandas.read_csv('corona_cases_belgium.csv', sep=';')

input_df
```

# Graphe simple des nombres de cas par jours

```python
plt.figure(figsize=(12, 5))
plt.plot(input_df['CASES'])
plt.xlabel('Date (index)')
plt.ylabel('Nombre de cas')
plt.title('Ã‰volution du nombre de cas par jour')
plt.show()
```

# Moyenne du nombre de cas par jour

```python
moyenne = input_df['CASES'].mean()
moyenne
```

# Graphe avec moyenne

```python
plt.figure(figsize=(12, 5))
plt.plot(input_df['CASES'])
plt.xlabel('Date (index)')
plt.ylabel('Nombre de cas')
plt.title('Ã‰volution du nombre de cas par jour')
plt.hlines(moyenne, 0, input_df.index.stop, color='orange')
plt.show()
```

# Histogramme

```python
plt.figure(figsize=(12, 5))
hist = plt.hist(input_df['CASES'], bins=50)
plt.xlabel('Nombre de cas')
plt.ylabel('Nombre de jours')
plt.title('Nombre de jours par nombre de cas')
plt.vlines(moyenne, 0, hist[0].max(), color='orange')
plt.show()
```

# Mode

```python
mode = hist[0].max()
mode
```

# Quartiles

```python
quartiles = input_df['CASES'].quantile([0.25, 0.5, 0.75])

plt.figure(figsize=(12, 5))
hist = plt.hist(input_df['CASES'], bins=50)
plt.xlabel('Nombre de cas')
plt.ylabel('Nombre de jours')
plt.title('Nombre de jours par nombre de cas')

plt.vlines(moyenne, 0, hist[0].max(), color='orange')
plt.vlines(quartiles, 0, hist[0].max(), color='red', linestyles='--')

plt.show()

quartiles
```
