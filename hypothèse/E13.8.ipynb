La dur√©e de vie moyenne d'un √©chantillon de 100 ampoules fluorescentes fabriqu√©es par une usine est estim√©e  √† 1570 heures avec un √©cart-type de 120 heures. Si ùúá est la dur√©e de vie moyenne de toutes les ampoules produites par l'usine, tester l'hypoth√®se ùúá=1600 heures avec l'hypoth√®se ùúá‚â†1600 heures, en choisissant un niveau de signification 
(a)  0.05 (b) de 0.01

- h0 : p = 1600
- h1 : p != 1600

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 1600
sigma = 120

n = 100
moyenne_observee = 1570
alpha = 0.05

sigma_moyenne = sigma / math.sqrt(n)

Z = (moyenne_observee - mu) / sigma_moyenne

Z_crit_right = norm.ppf(1 - alpha/2)
Z_crit_left = -Z_crit_right  # Pour un test bilat√©ral

print("Moyenne observ√©e =", moyenne_observee)
print("Sigma moyenne =", sigma_moyenne)
print("Valeur observ√©e Z =", Z)
print("Seuil critique =[", Z_crit_left, Z_crit_right, "]")

# Test bilat√©ral
if Z <= Z_crit_left or Z >= Z_crit_right:
    print("On rejette H0 : ")
else:
    print("On ne rejette pas H0 : ")


# --- Graphique ---
x = np.linspace(mu - 4*sigma_moyenne, mu + 4*sigma_moyenne, 1000)
y = norm.pdf(x, mu, sigma_moyenne)

plt.figure(figsize=(10,6))
plt.plot(x, y, label="Distribution sous H0")
plt.fill_between(x, 0, y, where=(x <= mu + Z_crit_left*sigma_moyenne), color='red', alpha=0.3, label="Zone critique")
plt.fill_between(x, 0, y, where=(x >= mu + Z_crit_right*sigma_moyenne), color='red', alpha=0.3)
plt.axvline(moyenne_observee, color='green', linestyle='--', label="Moyenne observ√©e")
plt.axvline(mu + Z_crit_left*sigma_moyenne, color='red', linestyle='--', label="Seuil critique")
plt.axvline(mu + Z_crit_right*sigma_moyenne, color='red', linestyle='--')
plt.title("Test d'hypoth√®se bilat√©ral")
plt.xlabel("Moyenne des dur√©es de vies (h)")
plt.ylabel("Densit√©")
plt.legend()
plt.grid(True)
plt.show()
```
