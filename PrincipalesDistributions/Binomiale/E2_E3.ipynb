```python
import scipy.stats
import matplotlib.pyplot as plt
```

```python
n = 10
p2 = 0.2
p3 = 0.5

def afficher_distributionBinomiale(n ,p):
    binomial = scipy.stats.binom(n, p)

    tous_les_cas_de_nombre_succes = range(0,n+1)
    list(tous_les_cas_de_nombre_succes)
    plt.figure(figsize=(12, 5))
    plt.scatter( tous_les_cas_de_nombre_succes, binomial.pmf(range(n+1)))
    plt.ylabel('Probabilit')
    plt.xlabel('Nombre de succs')
    plt.title(f'Distribution binomiale n={n}, p={p}')
    plt.show()

afficher_distributionBinomiale(n, p2)
afficher_distributionBinomiale(n, p3)
```
