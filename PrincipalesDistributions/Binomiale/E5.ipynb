```python
import scipy.stats
import matplotlib.pyplot as plt
import pandas as pd
```

# E5 5% des pièces produites par une machine sont défectueuses sur 4 pièces

- (a) = 1 défectueuse
- (b) = 0 défectueuse
- (c) <= 2 défectueuses

```python
p = 0.05
n = 4

df = pd.DataFrame()

df['NB_PIECE_CASSE'] = range(0, n + 1)
df['PROBA'] = scipy.stats.binom.pmf(df['NB_PIECE_CASSE'], n, p)
df
```

```python
print(f"1 pièce cassée : {df.loc[df['NB_PIECE_CASSE'] == 1, 'PROBA'].values[0]:.2f}")
print(f"0 pièce cassée : {df.loc[df['NB_PIECE_CASSE'] == 0, 'PROBA'].values[0]:.2f}")
print(f"au plus 2 pièce cassée : {df.loc[df['NB_PIECE_CASSE'] <= 2, 'PROBA'].sum():.4f}")
```
