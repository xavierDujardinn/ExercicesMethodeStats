```python
import scipy.stats
import matplotlib.pyplot as plt
import pandas as pd
```

# E4 : 2000 familles avec 4 garçons chacune

- (a) Combien peuvent s'attendre à avoir >= 1 garçon
- (b) = 2 garçons
- (c) <= 2 garçons
- (d) = 1 filles || = 2 filles
- (e) = 0 fomme 

```python
N = 2000
p = 0.5
n = 4

df = pd.DataFrame()

df['NB_GARCONS'] = range(0, n + 1)
df
```

```python
df['PROBA'] = scipy.stats.binom.pmf(df['NB_GARCONS'], n, p)
df
```

```python
df['NB_FAMILLES'] = (2000 * df['PROBA']).astype(int)
df
```

```python
print(f"Au moins 1 garçon : {N - df.loc[df['NB_GARCONS'] == 0, 'NB_FAMILLES'].values[0]} familles")
print(f"Exactement 2 garçons : {df.loc[df['NB_GARCONS'] == 2, 'NB_FAMILLES'].values[0]} familles")
print(f"Au plus 2 garçons : {df.loc[df['NB_GARCONS'] <= 2, 'NB_FAMILLES'].sum()} familles")
print(f"Une ou deux filles : {df.loc[(df['NB_GARCONS'] == 2) | (df['NB_GARCONS'] == 1), 'NB_FAMILLES'].sum()} familles")
print(f"Aucune fille : {df.loc[df['NB_GARCONS'] == 4, 'NB_FAMILLES'].values[0]} familles")
```
