```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.stats import norm
```

# E6 La taille moyenne de 500 élèves des petites classes d'un lycée est 151 cm et l'écart-type 15 cm
taille comprise entre (a) 120 et 155 cm, (b) 185 cm et plus

```python
n = 500

mu = 151
sigma = 15

df = pd.DataFrame()

df['TAILLE'] = range(0, 201)
df['PROBA'] = norm.pdf(df['TAILLE'], loc=mu, scale=sigma)

plt.figure(figsize=(12, 4))
plt.plot(df['TAILLE'], df['PROBA'])
plt.title("Distribution normale des tailles")
plt.xlabel("Taille (cm)")
plt.ylabel("Densité de probabilité")
plt.show()
```

```python
print(f"Nombre d'élèves mesurant entre 120 et 155 cm : {n*(norm.cdf(155, loc=mu, scale=sigma) - norm.cdf(120, loc=mu, scale=sigma))} élèves")
print(f"Nombre d'élèves mesurant au moins 185 cm : {n* (1-norm.cdf(185, loc=mu, scale=sigma))} élèves")
```

```python
plt.plot(df['TAILLE'],df['PROBA'],'-')
select_df = df[(df['TAILLE']>=120)&(df['TAILLE']<=155)]
plt.fill_between(select_df['TAILLE'], select_df['PROBA'], alpha=0.4, color='blue', label='120 cm à 155 cm')
plt.fill_between(df['TAILLE'], df['PROBA'], where=(df['TAILLE']>=185), alpha=0.4, color='red', label='au moins 185 cm')
plt.ylim(bottom=0)
plt.legend()
plt.xlabel('taille (cm)')
plt.ylabel('densité de probabilité')
plt.show()
```
