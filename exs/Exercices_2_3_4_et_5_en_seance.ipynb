```python
import scipy.stats
import matplotlib.pyplot as plt
```

```python
p = 0.5
n = 10
```

```python
binomial = scipy.stats.binom(n, p)
```

```python
binomial
```

```python
tous_les_cas_de_nombre_succes = range(0,n+1)
list(tous_les_cas_de_nombre_succes)
```

```python
binomial.pmf(tous_les_cas_de_nombre_succes)
```

```python
plt.scatter(tous_les_cas_de_nombre_succes, binomial.pmf(tous_les_cas_de_nombre_succes))
plt.show()
```
