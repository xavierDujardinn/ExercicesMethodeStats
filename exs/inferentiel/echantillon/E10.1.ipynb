# Population

```python
import pandas as pd
import itertools
import numpy as np

df = pd.DataFrame()

df['pop'] = [2,3,6,8,11]

echantillons = list(itertools.combinations(df['pop'], 2))

df
```

# Moyenne de la population

```python
mean_pop = df['pop'].mean()
mean_pop
```

# Ecart-Type de la population

```python
deltas = df['pop'] - mean_pop

variance = np.dot(deltas, deltas) / len(df['pop'])
np.sqrt(variance)
```

# Echantillons

```python
df_echantillons = pd.DataFrame(echantillons, columns=['valeur1', 'valeur2'])

df_echantillons['moyenne'] = df_echantillons.mean(axis=1)

df_echantillons
```

# Moyenne de la distribution des moyennes des échantillons

```python
mean_echantillon = df_echantillons['moyenne'].mean()
mean_echantillon
```

# Ecart-Type de la distribution des moyennes des échantillons

```python
df_echantillons['delta'] = df_echantillons['moyenne'] - mean_echantillon

variance_echantillon = np.dot(df_echantillons['delta'], df_echantillons['delta']) / len(df_echantillons['moyenne'])
np.sqrt(variance_echantillon)
```
