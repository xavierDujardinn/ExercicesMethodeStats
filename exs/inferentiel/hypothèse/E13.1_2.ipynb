```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import binom
```

```python
n = 100
p = 0.5

prob_rejet = binom.cdf(39, n, p) + (1 - binom.cdf(60, n, p))
prob_rejet
```

```python
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

plt.bar(x, y)
plt.axvline(0.4*n, color='red', linestyle='--')
plt.axvline(0.6*n, color='red', linestyle='--')
plt.title("Règle de décision pour le test de la pièce")
plt.xlabel("Nombre de faces")
plt.ylabel("Probabilité")
plt.show()
```
