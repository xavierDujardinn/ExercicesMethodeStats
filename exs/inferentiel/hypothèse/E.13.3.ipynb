```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import binom
```

```python
n = 100
x = 15
limite = 0.1
alpha = 0.01


# EXEMPLE PVALUE NE PAS PRENDRE EN COMPTE
p_value = 1 - binom.cdf(x - 1, n, limite)

print("p-value =", p_value)

if p_value < alpha:
    print("On rejette H0 : la machine doit être réparée.")
else:
    print("On ne rejette pas H0 : preuve insuffisante pour réparer.")
```

```python
import math
from scipy.stats import norm

moyenne_observee = x / n

sigma_moyenne_observee = math.sqrt(limite * (1 - limite) / n)

Z = (moyenne_observee - limite) / sigma_moyenne_observee

Z_crit = norm.ppf(1 - alpha)

print("Moyenne observée =", moyenne_observee)
print("Écart-type de la moyenne =", sigma_moyenne_observee)
print("Valeur observée Z =", Z)
print("Seuil critique =", Z_crit)

if Z >= Z_crit:
    print("On rejette H0 : la machine doit être réparée.")
else:
    print("On ne rejette pas H0 : preuve insuffisante pour réparer.")
```
