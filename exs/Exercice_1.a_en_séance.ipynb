```python
histogram = plt.hist(input_df['CASES'], bins=20)
plt.vlines(input_df['CASES'].mean(), 0, histogram[0].max(), color='orange')
plt.show()
```

```python
quartile = input_df['CASES'].quantile(q=[0.25, 0.50, 0.75])
quartile
```

```python
histogram = plt.hist(input_df['CASES'], bins=20)
plt.vlines(input_df['CASES'].mean(), 0, histogram[0].max(), color='orange')
for quantile_value in quartile.values:
    plt.vlines(quantile_value, 0, histogram[0].max(), color='red', linestyle='dotted' )
plt.show()
```
